# What was the win percentage, and the aggregated number of wins, losses, goals scored, and goals conceded for each team in each season across each league that they're associated with?

```python
# calculate the number of wins, losses, goals scored, and goals conceded and number of wins for each team in each season across each league
win_percentage = pd.read_sql("""select m.season,l.name as league ,t.team_long_name as team,
                                sum(m.home_team_goal) as scored_goals,
                                sum(m.away_team_goal) as conduct_goals,
                               
                                
                                
                                sum(case 
                                    when m.home_team_goal > m.away_team_goal then 1
                                    else 0
                                    end ) as "wins",
                                    
                                sum(case
                                    when  m.away_team_goal > m.home_team_goal then 1
                                    else 0
                                    end ) as "loose",
                                
                               sum (case 
                                    when m.away_team_goal = m.home_team_goal   then 1
                                    else 0
                                    end) as 'tie',
                                    
                                sum (case 
                                    when m.home_team_goal > m.away_team_goal then 1
                                    else 0
                                    end)* 100 /count(*)as win_percentage
                                    
                                    
                                from match as m
                                
                                inner join team as t
                                on t.team_api_id = m.home_team_api_id
                                
                                inner join league as l
                                using (country_id)
                                
                                
                                group by team ,season


                                union
                                
                                select m.season,l.name as league ,t.team_long_name as team,
                                sum(m.home_team_goal)as scored_goals,
                                sum(m.away_team_goal) as conduct_goals,
                                
                                
                                
                                sum(case 
                                    when m.away_team_goal > m.home_team_goal then 1
                                    else 0
                                    end ) as "wins",
                                    
                                sum(case
                                    when  m.home_team_goal > m.away_team_goal then 1
                                    else 0
                                    end ) as "loose",
                                
                               sum (case 
                                    when m.away_team_goal = m.home_team_goal   then 1
                                    else 0
                                    end) as 'tie',
                                    
                                sum (case 
                                    when m.away_team_goal > m.home_team_goal then 1
                                    else 0
                                    end)* 100 /count(*)as win_percentage
                                    
                                    
                                from match as m
                                
                                inner join team as t
                                on t.team_api_id = m.home_team_api_id
                                
                                inner join league as l
                                using (country_id)
                                
                                
                                group by team ,season 
                                order by win_percentage desc
                                limit 10;""",conn)

                               
win_percentage                               
```
